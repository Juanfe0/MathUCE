# **LÓGICA PROPOSICIONAL**

## **PROPOSICIÓN**

Una proposición es una afirmación que puede ser verdadera o falsa, pero no ambas. Estas afirmaciones son la base sobre la cual se construye la lógica proposicional.



## **OPERADORES LÓGICOS**

Los operadores lógicos principales son:

1. **Negación (¬)**
   - Invierte el valor de verdad de una proposición. Si una proposición es verdadera, su negación es falsa, y viceversa.

2. **Conjunción (∧)**
   - Representa la operación lógica "y". Una conjunción es verdadera solo si ambas proposiciones que la componen son verdaderas.

3. **Disyunción (∨)**
   - Representa la operación lógica "o". Una disyunción es verdadera si al menos una de las proposiciones que la componen es verdadera.

4. **Implicación (→)**
   - Representa la relación "si... entonces...". Una implicación es falsa solo si la primera proposición es verdadera y la segunda es falsa.

5. **Doble implicación (↔)**
   - Representa la relación "si y solo si". Una doble implicación es verdadera si ambas proposiciones tienen el mismo valor de verdad.

## **Cuantificador Universal (∀)**

Indica que una proposición es verdadera para todos los elementos de un conjunto. Se lee como "para todo".

## **Cuantificador Existencial (∃)**

Indica que existe al menos un elemento en el conjunto para el cual la proposición es verdadera. Se lee como "existe".

# ***TABLAS DE VERDAD***

## **¿QUE SON LAS TABLAS DE VERDAD?**

Una tabla de verdad es una representación sistemática de los posibles valores de verdad (verdadero o falso) de una proposición compuesta, en función de los valores de verdad de sus componentes individuales (variables proposicionales).

## **PROPOSITO Y FUNCION**

Las tablas de verdad se utilizan principalmente para:

- **Evaluar Combinaciones de Valores**: Permiten evaluar todas las combinaciones posibles de valores de verdad de las variables en una proposición compuesta.
- **Determinar la Validez Lógica**: Ayudan a determinar si una proposición compuesta es siempre verdadera (tautología), siempre falsa (contradicción) o si su valor de verdad depende de las variables involucradas.
- **Análisis de Circuitos Lógicos**: En la teoría de circuitos digitales, las tablas de verdad se utilizan para diseñar y analizar el comportamiento de los circuitos basados en lógica booleana

#***TEORIA DE CONJUNTOS***

## **CONJUNTOS:**

## **DEFINICION DE CONJUNTO**

Un conjunto es una colección de elementos distintos, que pueden ser números, letras u otros objetos. Los elementos de un conjunto no se repiten y pueden ser cualquier tipo de entidad claramente definida.

# **¿PARA QUE SIRVEN?**

- **Organización y Clasificación**: Los conjuntos permiten organizar y clasificar elementos de manera estructurada.
- **Teoría de Conjuntos**: La base para muchas ramas de las matemáticas, incluyendo álgebra, análisis y combinatoria.
- **Aplicaciones en Informática**: Uso en estructuras de datos, bases de datos y algoritmos.


## **CARACTERISTICAS DE LOS CONJUNTOS**

- **Elementos Distintos**: Los elementos de un conjunto son únicos; no hay duplicados.
- **No Importa el Orden**: El orden de los elementos en un conjunto no importa.
- **Definición Explícita**: Cada elemento del conjunto debe estar claramente definido y ser identificable.

## **OPERACIONES CON CONJUNTOS**


## **NOTACION**

Un conjunto se denota con llaves: {𝑎, 𝑏, 𝑐}

Conjunto vacío: {} o ∅.

---

## **OPERACIONES CON CONJUNTOS**

1. **Unión (A ∪ B)**
   - La unión de dos conjuntos A y B es el conjunto de todos los elementos que están en A, en B, o en ambos.
   - **Ejemplo**: {1, 2} ∪ {2, 3} = {1, 2, 3}.

2. **Intersección (A ∩ B)**
   - La intersección de dos conjuntos A y B es el conjunto de todos los elementos que están en ambos conjuntos.
   - **Ejemplo**: {1, 2} ∩ {2, 3} = {2}.

3. **Diferencia (A - B)**
   - La diferencia de dos conjuntos A y B es el conjunto de todos los elementos que están en A pero no en B.
   - **Ejemplo**: {1, 2} - {2, 3} = {1}.

4. **Complemento (A^c)**
   - El complemento de un conjunto A es el conjunto de todos los elementos que no están en A.

# **CONJUNTOS NUMERICOS**

## **¿QUE SON LOS CONJUNTOS NUMERICOS?**

Los conjuntos numéricos son agrupaciones de números que comparten características comunes. Son fundamentales en matemáticas y se utilizan para clasificar y entender diferentes tipos de números.

### **TIPOS DE CONJUNTOS NUMERICOS**

1. **Números Naturales (N)**
   - Conjunto de los números enteros no negativos: {0, 1, 2, 3, …, +∞}

2. **Números Enteros (Z)**
   - Conjunto de los números positivos, negativos y el cero: {−∞, …, −3, −2, −1, 0, 1, 2, 3, …, +∞}

3. **Números Racionales (Q)**
   - Conjunto de todos los números que pueden expresarse como una fracción de dos enteros: {m/n ; n ∈ Z, m ∈ N}, donde n no es cero.

4. **Números Reales (R)**
   - Conjunto de todos los números en la recta numérica, incluyendo números racionales e irracionales.

---

Por ende, N ⊆ Z ⊆ Q ⊆ R
