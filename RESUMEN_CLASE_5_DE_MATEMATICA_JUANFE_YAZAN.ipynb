# üìä‚ú® **Funciones de Utilidad en Econom√≠a**

## üåü **Introducci√≥n**

En econom√≠a, las funciones de utilidad son fundamentales para entender c√≥mo los individuos toman decisiones de consumo. Estas funciones miden el nivel de satisfacci√≥n o bienestar derivado del consumo de distintos bienes y servicios. La utilidad es una medida abstracta que permite cuantificar las preferencias de los consumidores.

## üìê **Definici√≥n**

Una **funci√≥n de utilidad** es una representaci√≥n matem√°tica que asigna un valor num√©rico a las diferentes combinaciones de bienes consumidos. Este valor refleja el nivel de satisfacci√≥n o utilidad que un consumidor obtiene.

## üßÆ **Ejemplo Pr√°ctico**

Para ilustrar el concepto, consideremos una funci√≥n de utilidad sencilla:

$[ U(x, y) = x \times y ]$

Donde:
- $(x)$ representa la cantidad del bien A consumido.
- $(y)$ representa la cantidad del bien B consumido.

## üîç **Interpretaci√≥n**

La funci√≥n anterior indica que la utilidad total es proporcional al producto de las cantidades de los dos bienes consumidos. Esto sugiere que consumir m√°s de ambos bienes aumenta la satisfacci√≥n del consumidor.

## üöÄ **Aplicaciones**

Las funciones de utilidad son utilizadas en modelos econ√≥micos para:
- **Determinar la cesta √≥ptima de bienes**: Identificar qu√© combinaciones de bienes maximizan la utilidad.
- **Estudiar el comportamiento del consumidor**: Comprender c√≥mo las decisiones de consumo cambian en respuesta a variaciones en precios e ingresos.
- **Analizar pol√≠ticas econ√≥micas**: Evaluar el impacto de pol√≠ticas sobre el bienestar de los individuos.




*   # üíßüí≠ **Funci√≥n de Utilidad con un Solo Bien**




## üåä **Ejemplo Simple: Agua**

Consideremos un ejemplo en el que el √∫nico bien disponible es el agua. La funci√≥n de utilidad en este caso se representa como \( u(x) \), donde \( x \) es la cantidad de agua consumida y \( u(x) \) es la utilidad obtenida.

### üìà **Relaci√≥n Agua-Utilidad**

$[ \text{Agua}(x) \rightarrow \text{Utilidad}(u(x)) ]$

Gr√°ficamente, esta relaci√≥n se puede representar con una curva que muestra c√≥mo la utilidad aumenta a medida que se consume m√°s agua. Sin embargo, la **utilidad marginal** decrece, lo que significa que cada unidad adicional de agua proporciona una utilidad menor que la anterior.

Por ejemplo, el primer vaso de agua puede proporcionar mucha satisfacci√≥n en un d√≠a caluroso, pero el tercer o cuarto vaso proporciona una satisfacci√≥n significativamente menor.

## üîç **Observaciones**

- **Utilidad marginal decreciente**: A medida que aumenta el consumo, la utilidad adicional obtenida disminuye.
- **Importancia del contexto**: El valor de cada unidad adicional depende de las circunstancias, como el clima o la sed.

Este comportamiento refleja una caracter√≠stica com√∫n en muchas funciones de utilidad y es clave para entender las decisiones de consumo en la vida real.




*   # üí≠‚ú® Funciones de Utilidad con Dos Bienes




## üåäü•ñ Introducci√≥n

La situaci√≥n se vuelve m√°s interesante cuando consideramos dos bienes, como agua y pan. Ahora, la funci√≥n de utilidad depende de ambas cantidades consumidas:

$[ u(x, y) ]$

Aqu√≠, $( x )$ es la cantidad de agua y $( y )$ es la cantidad de pan. La utilidad se representa en un gr√°fico tridimensional, donde los ejes $( x )$ e $( y )$ corresponden a las cantidades de los dos bienes, y el eje vertical representa la utilidad. Esta representaci√≥n gr√°fica muestra una **superficie de utilidad**.

## üåê Superficie de Utilidad

Imaginemos una superficie en forma de concha ac√∫stica, donde cada punto en la superficie representa un nivel de utilidad espec√≠fico asociado con una combinaci√≥n particular de $( x )$  $( y )$. Si solo miramos desde el lado de $( x )$ o $( y )$, veremos curvas similares a las que ten√≠amos en el caso unidimensional.

## üìä Ejemplos de Funciones de Utilidad

Ejemplos de funciones de utilidad que se ajustan a esta descripci√≥n incluyen:

- $( \log(x) )$
- $( \sqrt{x} )$

Ambas funciones tienen la caracter√≠stica de ser crecientes pero con una pendiente cada vez menor, reflejando el principio de **utilidad marginal decreciente**.

Estas funciones ayudan a modelar c√≥mo la satisfacci√≥n de un consumidor cambia con el consumo de diferentes combinaciones de bienes, proporcionando una herramienta √∫til para el an√°lisis econ√≥mico.




*   # üî¢‚ú® Notaci√≥n Matem√°tica de la Funci√≥n de Utilidad



1. **Definici√≥n de la Funci√≥n de Utilidad**:
   $[
   U: \mathbb{R}^2_+ \rightarrow \mathbb{R}^+
   ]$
   Esta notaci√≥n indica que la funci√≥n de utilidad $( U )$ toma como entrada un par de n√∫meros reales no negativos $(( \mathbb{R}^2_+ ))$ y produce un n√∫mero real no negativo $(( \mathbb{R}^+ ))$.

2. **Expresi√≥n de la Funci√≥n de Utilidad**:
   $[
   (x, y) \rightarrow u(x, y)
   ]$
   Aqu√≠, $( (x, y) $) son las cantidades de dos bienes, y $( u(x, y) )$ es el nivel de utilidad asociado con esas cantidades.

3. **Conjuntos de Nivel**:
   $[
   \{ (x, y) \in \mathbb{R}^2_+ : u(x, y) = 10 \}
   ]$
   Esto representa el conjunto de todos los pares $( (x, y) )$ en $( \mathbb{R}^2_+ )$ que producen una utilidad de 10. En el gr√°fico, estos conjuntos de nivel est√°n representados por **curvas de nivel** (isocuantas) en rojo.


# üìä ***Gr√°fica de la Funci√≥n de Utilidad en Python***





```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la funci√≥n de utilidad
def u(x, y):
    return np.log(x) + np.sqrt(y)

# Rango de valores para x e y
x = np.linspace(0.1, 5, 400)  # evitamos x=0 para evitar log(0)
y = np.linspace(0, 10, 400)
X, Y = np.meshgrid(x, y)
Z = u(X, Y)

# Creamos la figura
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Gr√°fico de la superficie de la funci√≥n de utilidad
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.8)

# Contornos en el plano
contours = ax.contour(X, Y, Z, levels=[1, 2, 3, 4, 5, 10], cmap='autumn', linestyles="solid")

# Etiquetas y t√≠tulo
ax.set_xlabel('Bien X')
ax.set_ylabel('Bien Y')
ax.set_zlabel('Utilidad U(x, y)')
ax.set_title('Superficie de la Funci√≥n de Utilidad y Conjuntos de Nivel')

# Mostrar el gr√°fico
plt.show()
```

# üìäüîç Funci√≥n de Utilidad: Conjuntos de Nivel y Preimagen

En el an√°lisis econ√≥mico, los conjuntos de nivel y la preimagen de una funci√≥n de utilidad son conceptos fundamentales que ayudan a entender c√≥mo las combinaciones de consumo afectan la satisfacci√≥n del consumidor.

## üìà Notaci√≥n Matem√°tica

### üîç Conjuntos de Nivel

Los conjuntos de nivel de una funci√≥n de utilidad $( u(x, y) )$ representan todas las combinaciones de bienes $( (x, y) )$ que proporcionan la misma utilidad.

$[ f^{-1}(\{10\}) = \{ (x, y) \in \mathbb{R}^2_+ : u(x, y) = 10 \} ]$

Para la funci√≥n $( u(x, y) = \sqrt{xy} )$, el conjunto de nivel para una utilidad de 10 se define como:

$[ \{ (x, y) \in \mathbb{R}^2_+ : \sqrt{xy} = 10 \} ]$

Simplificando, obtenemos:

$[ \{ (x, y) \in \mathbb{R}^2_+ : xy = 100 \} ]$

O, de manera equivalente:

$[ \{ (x, y) \in \mathbb{R}^2_+ : y = \frac{100}{x} \} ]$

### üåê Generalizaci√≥n para un Nivel de Utilidad \( u_0 \)

Para un nivel de utilidad \( u_0 \), el conjunto de nivel se generaliza como:

$[ f^{-1}(\{u_0\}) = \{ (x, y) \in \mathbb{R}^2_+ : y = \frac{u_0^2}{x} \} ]$

Esto indica que, independientemente del valor espec√≠fico de \( u_0 \), la relaci√≥n entre $( x )$ e \( y )$, se mantiene seg√∫n la ecuaci√≥n.

# üîÑ Curvas de Indiferencia de la Funci√≥n de Utilidad

En teor√≠a econ√≥mica, las curvas de indiferencia son representaciones gr√°ficas que muestran las combinaciones de dos bienes que proporcionan la misma utilidad total a un consumidor. Estas curvas son fundamentales para entender las preferencias y decisiones de consumo.

## üìö Definici√≥n

Una curva de indiferencia para una funci√≥n de utilidad $ u(x, y) $ es un conjunto de todas las combinaciones de cantidades $ (x, y) $ que proporcionan al consumidor la misma utilidad total. Matem√°ticamente, se define como:

$$ u(x, y) = \text{constante} $$

Esto implica que a lo largo de una curva de indiferencia, el consumidor es indiferente entre cualquier punto en la curva, ya que todos ofrecen la misma satisfacci√≥n.

## üìä Caracter√≠sticas

- **Declive**: Las curvas de indiferencia son descendentes de izquierda a derecha debido al principio de utilidad marginal decreciente. Esto significa que, para mantener la misma utilidad total, el consumidor debe recibir m√°s de un bien cuando tiene menos del otro.
  
- **Forma**: Las curvas de indiferencia pueden tener diferentes formas dependiendo de la funci√≥n de utilidad. Por ejemplo, si $ u(x, y) = \sqrt{x} + \sqrt{y} $, las curvas de indiferencia ser√°n c√≥ncavas hacia el origen.

## üìà Representaci√≥n Gr√°fica

En un plano $ xy $, cada curva de indiferencia representa una l√≠nea que conecta todos los puntos donde el consumidor se siente igualmente satisfecho en t√©rminos de utilidad. La combinaci√≥n exacta de bienes en cualquier punto de una curva de indiferencia depende de las preferencias del consumidor.

Las curvas de indiferencia son herramientas poderosas para analizar el comportamiento del consumidor y se utilizan extensamente en la teor√≠a econ√≥mica y la microeconom√≠a.


- üìä ***Curvas de indiferencia de la Funci√≥n de Utilidad***

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n de utilidad cuadr√°tica
def utility(x, y):
    return x**0.5 + y**0.5

# Valores de utilidad constante
utility_levels = [1, 2, 3]

# Crear un rango de valores para x e y
x = np.linspace(0.1, 5, 400)
y = np.linspace(0.1, 5, 400)
X, Y = np.meshgrid(x, y)

# Calcular las curvas de indiferencia
fig, ax = plt.subplots()

for u in utility_levels:
    Z = utility(X, Y)
    ax.contour(X, Y, Z, levels=[u], colors='b', linestyles='solid', linewidths=2)
    ax.text(0.1, 5, f'U={u}', fontsize=10)

ax.set_xlabel('Bien X')
ax.set_ylabel('Bien Y')
ax.set_title('Curvas de Indiferencia de la Funci√≥n de Utilidad')
ax.grid(True)
plt.xlim(0, 5)
plt.yl
```

# üíª ***Funci√≥n Sobreyectiva, Biyectiva, Inyectiva, Inversa y Compuesta***

En matem√°ticas, especialmente en teor√≠a de funciones, es fundamental entender diferentes tipos de funciones y sus propiedades. Aqu√≠ exploraremos las funciones sobreyectivas, biyectivas, inyectivas, as√≠ como sus inversas y composiciones.

- ## üéØ **Funci√≥n Inyectiva**

Una funci√≥n $ f: A \to B $ es inyectiva si cada elemento de $ A $ se mapea a un √∫nico elemento de $ B $. Matem√°ticamente, para todo $ x_1, x_2 \in A $, si $ f(x_1) = f(x_2) $, entonces $ x_1 = x_2 $.

- ## üîÑ **Funci√≥n Sobreyectiva**

Una funci√≥n $ f: A \to B $ es sobreyectiva si cada elemento de $ B $ tiene al menos un preimagen en $ A $. Esto significa que para todo $ y \in B $, existe al menos un $ x \in A $ tal que $ f(x) = y $.

- ## üîó **Funci√≥n Biyectiva**

Una funci√≥n $ f: A \to B $ es biyectiva si es tanto inyectiva como sobreyectiva. Esto implica que cada elemento de $ A $ se relaciona de manera √∫nica con un elemento de $ B $, y que cada elemento de $ B $ tiene exactamente una preimagen en $ A $.

- ## ‚Ü©Ô∏è Funci√≥n Inversa

Dada una funci√≥n biyectiva $ f: A \to B $, su funci√≥n inversa $ f^{-1}: B \to A $ est√° definida de manera que $ f^{-1}(f(x)) = x $ para todo $ x \in A $ y $ f(f^{-1}(y)) = y $ para todo $ y \in B $.

- ## üîÄ **Funci√≥n Compuesta**

La composici√≥n de dos funciones $ f: A \to B $ y $ g: B \to C $ es una nueva funci√≥n $ g \circ f: A \to C $, definida como $ (g \circ f)(x) = g(f(x)) $ para todo $ x \in A $.

Estos conceptos son esenciales para el estudio y la aplicaci√≥n de funciones en matem√°ticas, especialmente en √°lgebra y an√°lisis. La comprensi√≥n de funciones inyectivas, sobreyectivas, biyectivas, inversas y compuestas permite modelar y resolver una variedad de problemas matem√°ticos y aplicaciones pr√°cticas.


- ***Funci√≥n Sobreyectiva, Biyectiva, Inyectiva, Inversa y Compuesta, en codigo***

```python
# Definici√≥n de funciones

# Funci√≥n Sobreyectiva
def f_sobreyectiva(x):
    return x**2

# Funci√≥n Inyectiva
def f_inyectiva(x):
    return x

# Funci√≥n Biyectiva
def f_biyectiva(x):
    return x

# Funci√≥n Inversa
def inversa(x):
    return x**0.5  # Ra√≠z cuadrada positiva

# Funci√≥n Compuesta
def composicion(f, g, x):
    return f(g(x))

# Verificar sobreyectividad en un conjunto
def es_sobreyectiva(conjunto, funcion):
    imagenes = set()
    for elem in conjunto:
        imagen = funcion(elem)
        imagenes.add(imagen)
    return len(imagenes) == len(conjunto)

# Verificar inyectividad en un conjunto
def es_inyectiva(conjunto, funcion):
    imagenes = set()
    for elem in conjunto:
        imagen = funcion(elem)
        if imagen in imagenes:
            return False
        imagenes.add(imagen)
    return True

# Verificar biyectividad en un conjunto
def es_biyectiva(conjunto, funcion):
    return es_inyectiva(conjunto, funcion) and es_sobreyectiva(conjunto, funcion)

# Ejemplos de uso

# Conjunto de ejemplo
conjunto = set(range(-5, 6))  # Conjunto {-5, -4, ..., 4, 5}

# Funci√≥n Sobreyectiva
print("Funci√≥n Sobreyectiva:")
print(f"La funci√≥n f(x) = x^2 es sobreyectiva en el conjunto {-5, -4, ..., 4, 5}?")
print("Respuesta:", es_sobreyectiva(conjunto, f_sobreyectiva))

# Funci√≥n Inyectiva
print("\nFunci√≥n Inyectiva:")
print(f"La funci√≥n f(x) = x es inyectiva en el conjunto {-5, -4, ..., 4, 5}?")
print("Respuesta:", es_inyectiva(conjunto, f_inyectiva))

# Funci√≥n Biyectiva
print("\nFunci√≥n Biyectiva:")
print(f"La funci√≥n f(x) = x es biyectiva en el conjunto {-5, -4, ..., 4, 5}?")
print("Respuesta:", es_biyectiva(conjunto, f_biyectiva))

# Funci√≥n Inversa
print("\nFunci√≥n Inversa:")
valor_codominio = 4
print(f"El valor en el dominio que mapea a {valor_codominio} bajo la funci√≥n inversa es: {inversa(valor_codominio)}")

# Funci√≥n Compuesta
print("\nFunci√≥n Compuesta:")
def g(x):
    return x**2

def h(x):
    return x + 1

x = 2
print(f"El resultado de la composici√≥n g(h(x)) para x = {x} es: {composicion(g, h, x)}")
```

# ***Explicaci√≥n sobre Funciones Compuestas***

## üîë **Definiciones Iniciales:**

Se definen tres conjuntos:
- $ E $: Dominio inicial de la funci√≥n $ f $.
- $ F $: Codominio de $ f $ y dominio de $ g $.
- $ G $: Codominio de la funci√≥n $ g $.

Se definen dos funciones:
- $ f: E \rightarrow F $, donde $ f(x) = x^2 + 1 $.
- $ g: F \rightarrow G $, donde $ g(y) = \sqrt{y} $.

## üîë **Funci√≥n Compuesta:**

La funci√≥n compuesta $ g \circ f $ se define como:
$[ (g \circ f)(x) = g(f(x)) = \sqrt{f(x)} = \sqrt{x^2 + 1} ]$
Esta composici√≥n toma un elemento $ x $ de $ E $, aplica la funci√≥n $ f $ para obtener $ f(x) $ en $ F $, y luego aplica la funci√≥n $ g $ a $ f(x) $ para obtener $ g(f(x)) $ en $ G $.

## üîë **Ejemplo de Funci√≥n Compuesta:**

Se define la funci√≥n $ h(x) $:
$[ h(x) = \sqrt{x^2 + 1} ]$





*  ***Codigo de la Funcion Compuesta:***



```python
# Definici√≥n de funciones para la composici√≥n
def g(x):
    return x**2

def h(x):
    return x + 1

# Funci√≥n Compuesta
def composicion(f, g, x):
    return f(g(x))

# Ejemplo de uso de la funci√≥n compuesta g(h(x))
x = 2
resultado = composicion(g, h, x)
print(f"El resultado de la composici√≥n g(h(x)) para x = {x} es: {resultado}")
```
